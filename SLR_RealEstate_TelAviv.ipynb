```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing the dataset
```

```python
with open ('RealEstate_TLV.csv') as data:
#     print(data)
    dataset = pd.read_csv('RealEstate_TLV.csv')

# the relevant col is [year,price]
# year
y = dataset.iloc[:,1:2].values
# print(y)
# price
x = dataset.iloc[:,0:1].values
# print(x)
```

```python
# splitting the data into training set and test set
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# trainig the simple linear regression model -> trauning set
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# predicting the test results
```

```python
y_prediction = regressor.predict(x_test)
```

```python
# visualisation
```

```python
# showing the points
plt.scatter(x_train, y_train, color = 'red')
# showing the graph
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Real Estate price - Training set')
plt.xlabel('Years')
plt.ylabel('Price')
plt.show()
```

```python
# showing the points
plt.scatter(x_test, y_test, color = 'red')
# showing the graph
plt.plot(x_test, y_prediction, color = 'blue')
plt.title('Real Estate price - Test set')
plt.xlabel('Years')
plt.ylabel('Price')
plt.show()
```
